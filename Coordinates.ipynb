```python
import pandas as pd
import numpy as np

# Download the Trip Record Data
trip_record = pd.read_csv('~/Downloads/LSAProject/sampleYellowTripData.csv')

location_data = pd.read_csv('~/Downloads/LSAProject/taxi_zones.csv')
#location_data.head(3)


trip_record = trip_record.join(location_data.set_index('LocationID'), on=['PULocationID'])
trip_record.rename(columns={'X': 'PULocationX', 'Y': 'PULocationY'}, inplace=True)
trip_record.drop(['OBJECTID', 'Shape_Leng', 'Shape_Area', 'zone', 'borough'], axis=1, inplace=True)

trip_record = trip_record.join(location_data.set_index('LocationID'), on=['DOLocationID'])
trip_record.rename(columns={'X': 'DOLocationX', 'Y': 'DOLocationY'}, inplace=True)
trip_record.drop(['OBJECTID', 'Shape_Leng', 'Shape_Area', 'zone', 'borough'], axis=1, inplace=True)

trip_record['PULocationX'] = trip_record['PULocationX'].apply(lambda x: '{:.2f}'.format(x))
trip_record['DOLocationX'] = trip_record['DOLocationX'].apply(lambda x: '{:.2f}'.format(x))
print(trip_record)
```
